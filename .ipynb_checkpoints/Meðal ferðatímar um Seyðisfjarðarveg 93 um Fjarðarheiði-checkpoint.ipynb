# Áætlun raun-ferðatíma um Seyðisfjarðarveg 93 um Fjarðarheiði

Í skýrslum Rannsóknamiðstöðvar Háskólans á Akureyri er vægi styttingar ferðatíma til áætlunar greiðsluvilja mikil, en greiðsluvilji er hluti af arðsemismati jarðgangnakosta.

Fyrir Fjarðarheiði er stytting ferðatíma metin `6 mínútur` en miðar það við meðalhraða `70 km/klst`.<br>
Fyrir staðkunnuga virðast `70 km/klst` ofmat á hraða, einkum í vetrarfæri (nóvember-apríl), þar sem krappar beygjur, hálka og skyggni tefja ferðir oft verulega.

Vegna þessa var leitast við að fá nákvæmari ferðatíma með aðstoð notendagagna úr umferðarlíkani `Google Maps`.

Vegkaflinn sem um ræðir er:


Hugmyndin byggir á að kalla eftir ferðatímaspá, eitt ár fram í tímann milli bæjanna tveggja, og rannsaka árstíðabundnar sveiflur í ferðatímaspánni.

Til þess er kallað eftir ferðatímaspá með köllum í forritaskil (*e. Application Programming Interface, API*) `Google Maps`.

Á https://cloud.google.com/ má stofna notenda aðgang og fá `API_KEY` til að öðlast aðgang að þjónustunni. <br>
Þar sem greiða þarf sérstaklega ef fjöldi kalla í forritaskilin er verulegur, þarf að skrá greiðslukort.
Sökum þessa er lykill höfundar ekki opinberaður, heldur er honum varpað í `environment variable` í `terminal` áður en þessi stílabók (`jupyter-notebook`) er keyrð upp.

````python
import os, json, time
from datetime import datetime, timezone, timedelta
import requests

API_KEY = os.environ["GOOGLE_MAPS_API_KEY"]
URL = "https://routes.googleapis.com/directions/v2:computeRoutes"

EGILSSTADIR = {"location": {"latLng": {"latitude": 65.2669, "longitude": -14.3948}}}
SEYDISFJORDUR = {"location": {"latLng": {"latitude": 65.2603, "longitude": -14.0060}}}

FIELD_MASK = ",".join(["routes.duration",
                       "routes.staticDuration",
                       "routes.distanceMeters",
                       "routes.routeLabels",])
````

Við búum til `compute_route` fall, þar sem við skilgreinum `API-köllin`, og veljum `TRAFFIC_AWARE_OPTIMAL` í `routingPreference` til að fá áætlun sem byggir á raungögnum frá notendum (aðlöguð ferðatímaspá í krafti gagna Google Maps.

````python
def compute_route(origin, destination, departure_time_rfc3339=None):
    headers = {"Content-Type": "application/json",
               "X-Goog-Api-Key": API_KEY,
               "X-Goog-FieldMask": FIELD_MASK,}
    body = {"origin": origin,
            "destination": destination,
            "travelMode": "DRIVE",
            "routingPreference": "TRAFFIC_AWARE_OPTIMAL",}
    
    if departure_time_rfc3339:
        body["departureTime"] = departure_time_rfc3339

    r = requests.post(URL, headers=headers, json=body, timeout=30)
    
    if not r.ok:
        print("Status:", r.status_code)
        print(r.text)  
        r.raise_for_status()
    
    return r.json()
````

Nú getum við prófað að gera eitt kall í apann:

````python
def sample_once():
    ts = datetime.now(timezone.utc).isoformat()
    out = {
        "ts_utc": ts,
        "EGS_to_SEY": compute_route(EGILSSTADIR, SEYDISFJORDUR),
        "SEY_to_EGS": compute_route(SEYDISFJORDUR, EGILSSTADIR),
    }
    return out

sample = sample_once()
sample
````

OK, aðlögum þá fallið til að taka við tímastimpli, svo við getum óskað eftir ferðatímaáætlun í framtíðinni:

````python
def sample_from_timestamp(ts_rfc3339_utc: str):
    return {"ts_utc": ts_rfc3339_utc,
            "EGS_to_SEY": compute_route(EGILSSTADIR, SEYDISFJORDUR, departure_time_rfc3339=ts_rfc3339_utc),
            "SEY_to_EGS": compute_route(SEYDISFJORDUR, EGILSSTADIR, departure_time_rfc3339=ts_rfc3339_utc),}
````

Gerum eitt lítið sætt hjálparfall til að gera tímastimpil

````python
def create_timestamp(Y, M, D, h, m, s):
    return datetime(Y,M,D,h,m,s, tzinfo=timezone.utc).isoformat()
````

OK flott mál, þá er að græja tímastimpla fyrir heilt ár, raunar er einfaldast að gera það með `date_range` í `Pandas` svo við notum ekki einu sinni hjálparfallið góða...

````python
import pandas as pd

now = pd.Timestamp.utcnow()
start = now.ceil("H")
end = start + pd.DateOffset(years=1)

hours = pd.date_range(start, end, freq="H")
````

OK, nú er ég svolítið smeykur hvort við gætum nálgast að sprengja kvótann okkar, fyrir ókeypis köll í Google Maps API'ið...

Prófum að klukka hvað við náum raunhæft mörgum köllum í gegn á tímaeiningu, svo kreditkortareikningurinn verði ekki óvart alltof stór...

````python
def pick_metrics(resp):
    r0 = resp["routes"][0]
    return {"duration_s": pd.to_timedelta(r0["duration"]).total_seconds(),
            "static_duration_s": pd.to_timedelta(r0["staticDuration"]).total_seconds(),
            "distance_m": r0["distanceMeters"],}

rows = []

# Sömplum 10 færslur til að fá hugmynd um hraða...
for ts in hours[:10]:   
    
    stopwatch = time.time() # Ræsum skeiðklukku
    
    start = time
    ts_str = ts.to_pydatetime().isoformat()
    out = sample_from_timestamp(ts_str)

    e = pick_metrics(out["EGS_to_SEY"])
    s = pick_metrics(out["SEY_to_EGS"])

    rows.append({"departure_utc": ts_str,
                 "EGS_to_SEY_duration_s": e["duration_s"],
                 "EGS_to_SEY_static_s": e["static_duration_s"],
                 "SEY_to_EGS_duration_s": s["duration_s"],
                 "SEY_to_EGS_static_s": s["static_duration_s"],})
    # Stoppum og prentum skeiðklukku, fáum tíma á einni ítrun (tvö köll EGS->SEY og til baka)
    print(f"Duration: {time.time()-stopwatch:.4f} sec.")

df = pd.DataFrame(rows)
df.head()
````

Já ókei, við þurfum ekkert að hafa áhyggjur af neinum `3.000 calls/min` kvóta með þessar tölur.
Ef þetta væri í lágmarkinu, $0.4sek$ með hver tvö köll, værum við að tala um **best case** (worst case fyrir budduna):

$ \frac{2 köll}{0.4sek} \cdot 60 \frac{sek}{mín} = 300 \frac{köll}{mín}$

Jú jú, þá getum við farið að magnvinna þetta!

Instöllum `ipywidgets` með smá `inline magic` (%) `pip install'i`... þá getum við notað `tqdm` progress bar til að vakta magnvinnsluna, því við verðum örugglega einhvern klukkutíma að vinna þessu köll öllsömul...

Converta þessu snöggvast í `raw` format, má breyta aftur í `code` með því að velja selluna einfaldlega og ýta á `C`

Þá er komið að magnvinnslu, skilgreinum einn hjálparklasa, `RateMeter` sem getur hjálpað okkur að klippa á keyrsluna ef við erum að nálgast leyfilega, mínútulega, kvótann.

Sullum þessu svo út í `json` skrá sem við getum þá bara parse'að aftur inn í `DataFrame`, svo við séum ekki með allt í vinnsluminni og eigum á hættu á að tapa gögnunum sem við erum að sækja ef eitthvað klikkar hjá `apanum`.

````python
from pathlib import Path
from tqdm import tqdm 

OUT = Path("data/routes_sweep.jsonl")

def ts_to_rfc3339_z(ts: pd.Timestamp) -> str:
    return ts.isoformat().replace("+00:00", "Z")

class RateMeter:
    def __init__(self, window_seconds=60, min_elapsed_seconds=5):
        self.window = window_seconds
        self.min_elapsed = min_elapsed_seconds
        self.events = []  

    def add(self, n_requests: int):
        now = time.time()
        self.events.append((now, n_requests))
        cutoff = now - self.window
        while self.events and self.events[0][0] < cutoff:
            self.events.pop(0)

    def req_per_min(self) -> float:
        if len(self.events) < 2:
            return 0.0
        t0 = self.events[0][0]
        t1 = self.events[-1][0]
        elapsed = max(self.min_elapsed, t1 - t0)
        total = sum(n for _, n in self.events)
        return (total / elapsed) * 60.0

def run_sweep(hours: pd.DatetimeIndex,
              out_path: Path = OUT,
              quota_per_minute: int = 3000,
              stop_at_fraction: float = 0.8,
              warmup_samples: int = 10,):
    
    meter = RateMeter(window_seconds=60, min_elapsed_seconds=5)

    done = set()
    if out_path.exists():
        with out_path.open("r", encoding="utf-8") as f:
            for line in f:
                try:
                    obj = json.loads(line)
                    done.add(obj.get("departure_utc") or obj.get("ts_utc"))
                except Exception:
                    pass

    print(f"Output file: {out_path.resolve()}")
    print(f"Already completed: {len(done)} / {len(hours)} hours")
    print(f"Quota (ComputeRoutes): {quota_per_minute} req/min")
    threshold = quota_per_minute * stop_at_fraction
    print(f"Auto-stop threshold: {int(threshold)} req/min (after warmup: {warmup_samples} samples)")

    pbar = tqdm(hours, total=len(hours), desc="Sweeping hourly departures", unit="hour")
    samples = 0  

    for ts in pbar:
        dep = ts_to_rfc3339_z(ts)

        if dep in done:
            pbar.set_postfix(req_min=f"{meter.req_per_min():.1f}")
            continue

        t0 = time.time()
        out = sample_from_timestamp(dep)
        elapsed = time.time() - t0

        samples += 1
        meter.add(2) 
        req_min = meter.req_per_min()

        row = {"departure_utc": dep,
               "ts_utc": out.get("ts_utc", dep),
               "EGS_to_SEY": out["EGS_to_SEY"],
               "SEY_to_EGS": out["SEY_to_EGS"],
               "elapsed_s": elapsed,}
        
        with out_path.open("a", encoding="utf-8") as f:
            f.write(json.dumps(row) + "\n")

        pbar.set_postfix(req_min=f"{req_min:.1f}", last_s=f"{elapsed:.3f}", samples=samples)

        if samples >= warmup_samples and req_min > threshold:
            print(f"\nStopping: rolling rate {req_min:.1f} req/min exceeded "
                  f"{threshold:.0f} req/min threshold after {samples} samples.")
            break

    print("Sweep loop ended.")

run_sweep(hours)
````

**Geggt!**<br>
Importum núna `json'inunm`

````python
IN = Path("data/routes_sweep.jsonl")

rows = []

with IN.open("r", encoding="utf-8") as f:
    for i, line in enumerate(f, 1):
        try:
            obj = json.loads(line)

            dep = obj["departure_utc"]
            e = pick_metrics(obj["EGS_to_SEY"])
            s = pick_metrics(obj["SEY_to_EGS"])

            rows.append({"departure_utc": pd.to_datetime(dep, utc=True),
                         # Egilsstaðir → Seyðisfjörður
                         "EGS_to_SEY_duration_s": e["duration_s"],
                         "EGS_to_SEY_static_s": e["static_duration_s"],
                         "EGS_to_SEY_distance_m": e["distance_m"],
                         # Seyðisfjörður → Egilsstaðir
                         "SEY_to_EGS_duration_s": s["duration_s"],
                         "SEY_to_EGS_static_s": s["static_duration_s"],
                         "SEY_to_EGS_distance_m": s["distance_m"],})

        except Exception as ex:
            print(f"Skipping malformed line {i}: {ex}")

df = pd.DataFrame(rows).sort_values("departure_utc").reset_index(drop=True)
````

````python
df.describe()
````

````python
# traffic penalty vs historical baseline
df["EGS_to_SEY_delta_s"] = df["EGS_to_SEY_duration_s"] - df["EGS_to_SEY_static_s"]
df["SEY_to_EGS_delta_s"] = df["SEY_to_EGS_duration_s"] - df["SEY_to_EGS_static_s"]

# mean of both directions (often useful)
df["mean_duration_s"] = (    df["EGS_to_SEY_duration_s"] + df["SEY_to_EGS_duration_s"]) / 2.0
df["mean_static_s"] = (    df["EGS_to_SEY_static_s"] + df["SEY_to_EGS_static_s"]) / 2.0
df["mean_delta_s"] = df["mean_duration_s"] - df["mean_static_s"]
````
